```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/fraudTrain.csv')
print("Dataset loaded successfully")
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Drop rows with missing target value
df = df[df['is_fraud'].notna()].copy()

# Fill missing numeric values with median
num_cols = ['long', 'city_pop', 'unix_time', 'merch_lat', 'merch_long']
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Fill missing categorical value with mode
df['job'] = df['job'].fillna(df['job'].mode()[0])

# Fill missing date with placeholder
df['dob'] = df['dob'].fillna('1900-01-01')

# Fill missing transaction number with "Unknown"
df['trans_num'] = df['trans_num'].fillna('Unknown')
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("\nFraud vs Safe counts:")
fraud_counts = df['is_fraud'].value_counts()
print(fraud_counts)

print("\nFraud vs Safe percentages:")
print(df['is_fraud'].value_counts(normalize=True) * 100)
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='is_fraud', data=df)
plt.title('Distribution of Fraudulent vs. Non-Fraudulent Transactions')
plt.xlabel('Is Fraudulent')
plt.ylabel('Count')
plt.xticks([0, 1], ['Non-Fraudulent', 'Fraudulent'])
plt.show()
```

```python
# Separate fraud and safe transactions
fraud_df = df[df['is_fraud'] == 1]
safe_df = df[df['is_fraud'] == 0]

print("Original safe count:", len(safe_df))
print("Original fraud count:", len(fraud_df))

# Randomly sample 20,000 safe transactions
safe_sampled = safe_df.sample(n=20000, random_state=42)

# Combine fraud + sampled safe data
balanced_df = pd.concat([fraud_df, safe_sampled])

# Shuffle the combined data
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

print("\nBalanced Dataset Shape:", balanced_df.shape)
print(balanced_df['is_fraud'].value_counts())

# Visualize the new distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='is_fraud', data=balanced_df)
plt.title('Balanced Fraud vs Non-Fraud Transactions')
plt.xlabel('Is Fraudulent')
plt.ylabel('Count')
plt.xticks([0, 1], ['Non-Fraudulent', 'Fraudulent'])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Separate features & target
X_bal = balanced_df.drop(columns=['is_fraud'])
y_bal = balanced_df['is_fraud']

# Train-test split (balanced data)
X_train, X_test, y_train, y_test = train_test_split(
    X_bal, y_bal, test_size=0.2, stratify=y_bal, random_state=42
)

# Identify numeric & categorical features
numeric_features = X_train.select_dtypes(include=np.number).columns.tolist()
if 'Unnamed: 0' in numeric_features:
    numeric_features.remove('Unnamed: 0')
categorical_features = X_train.select_dtypes(include='object').columns.tolist()

print("Numeric features:", numeric_features)
print("Categorical features:", categorical_features)

# Transformers
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="constant", fill_value="missing")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=True))
])

# Preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ],
    remainder='passthrough'
)

# Apply preprocessing
X_train_pre = preprocessor.fit_transform(X_train)
X_test_pre = preprocessor.transform(X_test)

print("✅ Preprocessing complete")
print("Preprocessed Train Shape:", X_train_pre.shape)
print("Preprocessed Test Shape:", X_test_pre.shape)
```

```python
from sklearn.decomposition import TruncatedSVD

# TruncatedSVD for sparse data (like one-hot encoded)
n_components = 50
svd = TruncatedSVD(n_components=n_components, random_state=42)

X_train_pca = svd.fit_transform(X_train_pre)
X_test_pca = svd.transform(X_test_pre)

print("Original features:", X_train_pre.shape[1])
print("Reduced features:", X_train_pca.shape[1])

# Explained variance
explained_var_ratio = svd.explained_variance_ratio_.cumsum()

plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_var_ratio) + 1), explained_var_ratio, marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("TruncatedSVD - Cumulative Explained Variance")
plt.grid(True)
plt.show()

print(f"Explained Variance with {n_components} components: {explained_var_ratio[-1]:.2%}")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve, auc
import seaborn as sns
import matplotlib.pyplot as plt

# --- Scale features ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_pca)
X_test_scaled = scaler.transform(X_test_pca)

# --- Train Logistic Regression ---
log_reg = LogisticRegression(max_iter=5000, random_state=42)
log_reg.fit(X_train_scaled, y_train)

# --- Predictions ---
y_pred_lr = log_reg.predict(X_test_scaled)
y_prob_lr = log_reg.predict_proba(X_test_scaled)[:,1]

# --- Metrics Table ---
accuracy = accuracy_score(y_test, y_pred_lr)
roc_auc_val = roc_auc_score(y_test, y_prob_lr)
report = classification_report(y_test, y_pred_lr, output_dict=True)

# Get class keys dynamically (skip summary keys)
class_keys = [k for k in report.keys() if k not in ("accuracy", "macro avg", "weighted avg")]

# Prepare table rows (without support column)
rows = []
rows.append(["Accuracy", f"{accuracy:.3f}", "-", "-", "-"])
rows.append(["ROC-AUC", f"{roc_auc_val:.2f}", "-", "-", "-"])
for label in class_keys:
    rows.append([
        label,
        "-",
        f"{report[label]['precision']:.3f}",
        f"{report[label]['recall']:.3f}",
        f"{report[label]['f1-score']:.3f}"
    ])
columns = ["", "Value", "precision", "recall", "f1-score"]

# --- Confusion Matrix ---
cm = confusion_matrix(y_test, y_pred_lr)

# --- ROC Curve ---
fpr, tpr, _ = roc_curve(y_test, y_prob_lr)
roc_auc_val = auc(fpr, tpr)

# --- Plot Layout ---
fig, axes = plt.subplots(1, 3, figsize=(22,5))

# 1. Metrics Table
axes[0].axis("off")
table = axes[0].table(cellText=rows,
                      colLabels=columns,
                      cellLoc="center",
                      loc="center")
table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1.2, 1.2)
axes[0].set_title("Model Metrics", fontsize=14)

# 2. Confusion Matrix
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False, ax=axes[1])
axes[1].set_title("Confusion Matrix")
axes[1].set_xlabel("Predicted")
axes[1].set_ylabel("Actual")

# 3. ROC Curve
axes[2].plot(fpr, tpr, color="blue", label=f"AUC = {roc_auc_val:.2f}")
axes[2].plot([0,1],[0,1], linestyle="--", color="gray")
axes[2].set_title("ROC Curve")
axes[2].set_xlabel("False Positive Rate")
axes[2].set_ylabel("True Positive Rate")
axes[2].legend()

plt.tight_layout()
plt.show()
```

```python
# --- Model Comparison: Metrics Table & ROC Curves ---
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve, auc
from xgboost import XGBClassifier
import lightgbm as lgb
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# --- Initialize models ---
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Linear Regression": LinearRegression(),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    "LightGBM": lgb.LGBMClassifier()
}

# --- Store results ---
results = {}

# --- Train and Evaluate ---
for name, model in models.items():
    model.fit(X_train_pca, y_train)

    # Predictions
    if name == "Linear Regression":
        # Convert regression output to binary for comparison
        y_pred = (model.predict(X_test_pca) >= 0.5).astype(int)
        y_prob = model.predict(X_test_pca)
    else:
        y_pred = model.predict(X_test_pca)
        y_prob = model.predict_proba(X_test_pca)[:,1]

    # Metrics
    acc = accuracy_score(y_test, y_pred)
    roc_auc_val = roc_auc_score(y_test, y_prob)

    # Store
    results[name] = {
        "accuracy": acc,
        "roc_auc": roc_auc_val,
        "fpr": roc_curve(y_test, y_prob)[0],
        "tpr": roc_curve(y_test, y_prob)[1]
    }

# --- Prepare Table ---
rows = []
columns = ["Model", "Accuracy", "ROC-AUC"]
for model_name, metrics in results.items():
    rows.append([model_name, f"{metrics['accuracy']:.3f}", f"{metrics['roc_auc']:.3f}"])

# --- Plot Layout ---
fig, axes = plt.subplots(1, 2, figsize=(18,6))

# 1. Metrics Table
axes[0].axis("off")
table = axes[0].table(cellText=rows,
                      colLabels=columns,
                      cellLoc="center",
                      loc="center")
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.5, 1.5)
axes[0].set_title("Model Comparison Metrics", fontsize=14)

# 2. ROC Curves
for name, metrics in results.items():
    axes[1].plot(metrics['fpr'], metrics['tpr'], lw=2, label=f"{name} (AUC = {metrics['roc_auc']:.2f})")
axes[1].plot([0,1], [0,1], color='gray', linestyle='--')
axes[1].set_title("ROC Curve Comparison")
axes[1].set_xlabel("False Positive Rate")
axes[1].set_ylabel("True Positive Rate")
axes[1].legend(loc="lower right")

plt.tight_layout()
plt.show()
```
