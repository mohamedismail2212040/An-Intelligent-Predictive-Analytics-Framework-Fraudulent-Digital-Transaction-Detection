```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/fraudTrain.csv')
print("Dataset loaded successfully")
```

```python
# Drop rows with missing target value
df = df[df['is_fraud'].notna()].copy()

# Fill missing numeric values with median
num_cols = ['long', 'city_pop', 'unix_time', 'merch_lat', 'merch_long']
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Fill missing categorical value with mode
df['job'] = df['job'].fillna(df['job'].mode()[0])

# Fill missing date with placeholder
df['dob'] = df['dob'].fillna('1900-01-01')

# Fill missing transaction number with "Unknown"
df['trans_num'] = df['trans_num'].fillna('Unknown')
```

```python
# Separate fraud and safe transactions
fraud_df = df[df['is_fraud'] == 1]
safe_df = df[df['is_fraud'] == 0]

print("Original safe count:", len(safe_df))
print("Original fraud count:", len(fraud_df))

# Randomly sample 20,000 safe transactions
safe_sampled = safe_df.sample(n=20000, random_state=42)

# Combine fraud + sampled safe data
balanced_df = pd.concat([fraud_df, safe_sampled])

# Shuffle the combined data
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

print("\nBalanced Dataset Shape:", balanced_df.shape)
print(balanced_df['is_fraud'].value_counts())

# Visualize the new distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='is_fraud', data=balanced_df)
plt.title('Balanced Fraud vs Non-Fraud Transactions')
plt.xlabel('Is Fraudulent')
plt.ylabel('Count')
plt.xticks([0, 1], ['Non-Fraudulent', 'Fraudulent'])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Separate features & target
X_bal = balanced_df.drop(columns=['is_fraud'])
y_bal = balanced_df['is_fraud']

# Train-test split (balanced data)
X_train, X_test, y_train, y_test = train_test_split(
    X_bal, y_bal, test_size=0.2, stratify=y_bal, random_state=42
)

# Identify numeric & categorical features
numeric_features = X_train.select_dtypes(include=np.number).columns.tolist()
if 'Unnamed: 0' in numeric_features:
    numeric_features.remove('Unnamed: 0')
categorical_features = X_train.select_dtypes(include='object').columns.tolist()

print("Numeric features:", numeric_features)
print("Categorical features:", categorical_features)

# Transformers
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="constant", fill_value="missing")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=True))
])

# Preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ],
    remainder='passthrough'
)

# Apply preprocessing
X_train_pre = preprocessor.fit_transform(X_train)
X_test_pre = preprocessor.transform(X_test)

print("✅ Preprocessing complete")
print("Preprocessed Train Shape:", X_train_pre.shape)
print("Preprocessed Test Shape:", X_test_pre.shape)
```

```python
from sklearn.decomposition import TruncatedSVD

# TruncatedSVD for sparse data (like one-hot encoded)
n_components = 50
svd = TruncatedSVD(n_components=n_components, random_state=42)

X_train_pca = svd.fit_transform(X_train_pre)
X_test_pca = svd.transform(X_test_pre)

print("Original features:", X_train_pre.shape[1])
print("Reduced features:", X_train_pca.shape[1])

# Explained variance
explained_var_ratio = svd.explained_variance_ratio_.cumsum()

plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_var_ratio) + 1), explained_var_ratio, marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("TruncatedSVD - Cumulative Explained Variance")
plt.grid(True)
plt.show()

print(f"Explained Variance with {n_components} components: {explained_var_ratio[-1]:.2%}")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

# --- Scale features ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_pca)
X_test_scaled = scaler.transform(X_test_pca)

# --- Reshape for LSTM (samples, timesteps, features) ---
X_train_lstm = X_train_scaled.reshape(X_train_scaled.shape[0], 1, X_train_scaled.shape[1])
X_test_lstm = X_test_scaled.reshape(X_test_scaled.shape[0], 1, X_test_scaled.shape[1])

# --- Build LSTM Model ---
lstm_model = Sequential()
lstm_model.add(LSTM(64, input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2]), return_sequences=False))
lstm_model.add(Dropout(0.3))
lstm_model.add(Dense(32, activation='relu'))
lstm_model.add(Dense(1, activation='sigmoid'))

lstm_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# --- Train Model ---
history = lstm_model.fit(X_train_lstm, y_train, epochs=50, batch_size=64, validation_split=0.2, verbose=2)

# --- Predictions ---
y_prob_lstm = lstm_model.predict(X_test_lstm).ravel()
y_pred_lstm = (y_prob_lstm > 0.5).astype(int)

# --- Metrics Table ---
accuracy = accuracy_score(y_test, y_pred_lstm)
roc_auc_val = roc_auc_score(y_test, y_prob_lstm)
report = classification_report(y_test, y_pred_lstm, output_dict=True)

class_keys = [k for k in report.keys() if k not in ("accuracy", "macro avg", "weighted avg")]

rows = []
rows.append(["Accuracy", f"{accuracy:.3f}", "-", "-", "-"])
rows.append(["ROC-AUC", f"{roc_auc_val:.2f}", "-", "-", "-"])
for label in class_keys:
    rows.append([
        label,
        "-",
        f"{report[label]['precision']:.3f}",
        f"{report[label]['recall']:.3f}",
        f"{report[label]['f1-score']:.3f}"
    ])
columns = ["", "Value", "precision", "recall", "f1-score"]

# --- Confusion Matrix ---
cm = confusion_matrix(y_test, y_pred_lstm)

# --- ROC Curve ---
fpr, tpr, _ = roc_curve(y_test, y_prob_lstm)
roc_auc_val = auc(fpr, tpr)

# --- Plot Layout ---
fig, axes = plt.subplots(1, 3, figsize=(22,5))

# 1. Metrics Table
axes[0].axis("off")
table = axes[0].table(cellText=rows,
                      colLabels=columns,
                      cellLoc="center",
                      loc="center")
table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1.2, 1.2)
axes[0].set_title("LSTM Model Metrics", fontsize=14)

# 2. Confusion Matrix
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False, ax=axes[1])
axes[1].set_title("Confusion Matrix")
axes[1].set_xlabel("Predicted")
axes[1].set_ylabel("Actual")

# 3. ROC Curve
axes[2].plot(fpr, tpr, color="blue", label=f"AUC = {roc_auc_val:.2f}")
axes[2].plot([0,1],[0,1], linestyle="--", color="gray")
axes[2].set_title("ROC Curve")
axes[2].set_xlabel("False Positive Rate")
axes[2].set_ylabel("True Positive Rate")
axes[2].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout, Layer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

# --- Custom Attention Layer ---
class Attention(Layer):
    def __init__(self, **kwargs):
        super(Attention, self).__init__(**kwargs)

    def build(self, input_shape):
        self.W = self.add_weight(name="att_weight", shape=(input_shape[-1], input_shape[-1]),
                                 initializer="random_normal", trainable=True)
        self.b = self.add_weight(name="att_bias", shape=(input_shape[-1],),
                                 initializer="zeros", trainable=True)
        self.u = self.add_weight(name="att_u", shape=(input_shape[-1], 1),
                                 initializer="random_normal", trainable=True)
        super(Attention, self).build(input_shape)

    def call(self, inputs):
        u_it = tf.tanh(tf.tensordot(inputs, self.W, axes=1) + self.b)
        a_it = tf.nn.softmax(tf.tensordot(u_it, self.u, axes=1), axis=1)
        output = tf.reduce_sum(inputs * a_it, axis=1)
        return output

# --- Scale features ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_pca)
X_test_scaled = scaler.transform(X_test_pca)

# --- Reshape for LSTM (samples, timesteps, features) ---
X_train_lstm = X_train_scaled.reshape(X_train_scaled.shape[0], 1, X_train_scaled.shape[1])
X_test_lstm = X_test_scaled.reshape(X_test_scaled.shape[0], 1, X_test_scaled.shape[1])

# --- Build Attention LSTM Model ---
inputs = Input(shape=(X_train_lstm.shape[1], X_train_lstm.shape[2]))
x = LSTM(64, return_sequences=True)(inputs)
x = Dropout(0.3)(x)
x = Attention()(x)
x = Dense(32, activation='relu')(x)
outputs = Dense(1, activation='sigmoid')(x)

model = Model(inputs=inputs, outputs=outputs)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# --- Train Model ---
history = model.fit(X_train_lstm, y_train, epochs=50, batch_size=64, validation_split=0.2, verbose=2)

# --- Predictions ---
y_prob_att_lstm = model.predict(X_test_lstm).ravel()
y_pred_att_lstm = (y_prob_att_lstm > 0.5).astype(int)

# --- Metrics Table ---
accuracy = accuracy_score(y_test, y_pred_att_lstm)
roc_auc_val = roc_auc_score(y_test, y_prob_att_lstm)
report = classification_report(y_test, y_pred_att_lstm, output_dict=True)

class_keys = [k for k in report.keys() if k not in ("accuracy", "macro avg", "weighted avg")]

rows = []
rows.append(["Accuracy", f"{accuracy:.3f}", "-", "-", "-"])
rows.append(["ROC-AUC", f"{roc_auc_val:.2f}", "-", "-", "-"])
for label in class_keys:
    rows.append([
        label,
        "-",
        f"{report[label]['precision']:.3f}",
        f"{report[label]['recall']:.3f}",
        f"{report[label]['f1-score']:.3f}"
    ])
columns = ["", "Value", "precision", "recall", "f1-score"]

# --- Confusion Matrix ---
cm = confusion_matrix(y_test, y_pred_att_lstm)

# --- ROC Curve ---
fpr, tpr, _ = roc_curve(y_test, y_prob_att_lstm)
roc_auc_val = auc(fpr, tpr)

# --- Plot Layout ---
fig, axes = plt.subplots(1, 3, figsize=(22,5))

# 1. Metrics Table
axes[0].axis("off")
table = axes[0].table(cellText=rows,
                      colLabels=columns,
                      cellLoc="center",
                      loc="center")
table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1.2, 1.2)
axes[0].set_title("Attention LSTM Model Metrics", fontsize=14)

# 2. Confusion Matrix
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False, ax=axes[1])
axes[1].set_title("Confusion Matrix")
axes[1].set_xlabel("Predicted")
axes[1].set_ylabel("Actual")

# 3. ROC Curve
axes[2].plot(fpr, tpr, color="blue", label=f"AUC = {roc_auc_val:.2f}")
axes[2].plot([0,1],[0,1], linestyle="--", color="gray")
axes[2].set_title("ROC Curve")
axes[2].set_xlabel("False Positive Rate")
axes[2].set_ylabel("True Positive Rate")
axes[2].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

# --- Scale features ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_pca)
X_test_scaled = scaler.transform(X_test_pca)

# --- Reshape for GRU ---
X_train_gru = X_train_scaled.reshape(X_train_scaled.shape[0], 1, X_train_scaled.shape[1])
X_test_gru = X_test_scaled.reshape(X_test_scaled.shape[0], 1, X_test_scaled.shape[1])

# --- Build GRU Model ---
gru_model = Sequential()
gru_model.add(GRU(64, input_shape=(X_train_gru.shape[1], X_train_gru.shape[2])))
gru_model.add(Dropout(0.3))
gru_model.add(Dense(32, activation='relu'))
gru_model.add(Dense(1, activation='sigmoid'))

gru_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# --- Train Model ---
history = gru_model.fit(X_train_gru, y_train, epochs=50, batch_size=64, validation_split=0.2, verbose=2)

# --- Predictions ---
y_prob_gru = gru_model.predict(X_test_gru).ravel()
y_pred_gru = (y_prob_gru > 0.5).astype(int)

# --- Metrics, Confusion Matrix, ROC ---
accuracy = accuracy_score(y_test, y_pred_gru)
roc_auc_val = roc_auc_score(y_test, y_prob_gru)
report = classification_report(y_test, y_pred_gru, output_dict=True)
class_keys = [k for k in report.keys() if k not in ("accuracy", "macro avg", "weighted avg")]

rows = []
rows.append(["Accuracy", f"{accuracy:.3f}", "-", "-", "-"])
rows.append(["ROC-AUC", f"{roc_auc_val:.2f}", "-", "-", "-"])
for label in class_keys:
    rows.append([
        label,
        "-",
        f"{report[label]['precision']:.3f}",
        f"{report[label]['recall']:.3f}",
        f"{report[label]['f1-score']:.3f}"
    ])
columns = ["", "Value", "precision", "recall", "f1-score"]

cm = confusion_matrix(y_test, y_pred_gru)
fpr, tpr, _ = roc_curve(y_test, y_prob_gru)
roc_auc_val = auc(fpr, tpr)

fig, axes = plt.subplots(1, 3, figsize=(22,5))

axes[0].axis("off")
table = axes[0].table(cellText=rows, colLabels=columns, cellLoc="center", loc="center")
table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1.2, 1.2)
axes[0].set_title("GRU Model Metrics", fontsize=14)

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False, ax=axes[1])
axes[1].set_title("Confusion Matrix")
axes[1].set_xlabel("Predicted")
axes[1].set_ylabel("Actual")

axes[2].plot(fpr, tpr, color="blue", label=f"AUC = {roc_auc_val:.2f}")
axes[2].plot([0,1],[0,1], linestyle="--", color="gray")
axes[2].set_title("ROC Curve")
axes[2].set_xlabel("False Positive Rate")
axes[2].set_ylabel("True Positive Rate")
axes[2].legend()

plt.tight_layout()
plt.show()
```

```python
# --- Model Comparison: DL Models Metrics Table & ROC Curves (LSTM, Attention-LSTM, GRU) ---
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve

# --- Assuming predictions & probabilities from your trained DL models ---
# y_test: true labels
# y_pred_lstm, y_prob_lstm
# y_pred_att_lstm, y_prob_att_lstm
# y_pred_gru, y_prob_gru

# --- Initialize model results dictionary ---
results_dl = {
    "LSTM": {"y_pred": y_pred_lstm, "y_prob": y_prob_lstm},
    "Attention-LSTM": {"y_pred": y_pred_att_lstm, "y_prob": y_prob_att_lstm},
    "GRU": {"y_pred": y_pred_gru, "y_prob": y_prob_gru}
}

# --- Compute metrics ---
for name, data in results_dl.items():
    y_pred = data["y_pred"]
    y_prob = data["y_prob"]
    acc = accuracy_score(y_test, y_pred)
    roc_auc_val = roc_auc_score(y_test, y_prob)
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    results_dl[name]["accuracy"] = acc
    results_dl[name]["roc_auc"] = roc_auc_val
    results_dl[name]["fpr"] = fpr
    results_dl[name]["tpr"] = tpr
    print(f"{name}: Accuracy = {acc:.4f}, ROC-AUC = {roc_auc_val:.4f}")

# --- Prepare Table ---
rows = []
columns = ["Model", "Accuracy", "ROC-AUC"]
for model_name, metrics in results_dl.items():
    rows.append([model_name, f"{metrics['accuracy']:.3f}", f"{metrics['roc_auc']:.3f}"])

# --- Plot Layout ---
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# 1️⃣ Metrics Table
axes[0].axis("off")
table = axes[0].table(cellText=rows,
                      colLabels=columns,
                      cellLoc="center",
                      loc="center")
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.5, 1.5)
axes[0].set_title("Deep Learning Model Comparison Metrics", fontsize=14)

# 2️⃣ ROC Curves
colors = {
    "LSTM": "orange",
    "Attention-LSTM": "purple",
    "GRU": "red"
}

for name, metrics in results_dl.items():
    axes[1].plot(metrics["fpr"], metrics["tpr"], lw=2,
                 color=colors[name], label=f"{name} (AUC = {metrics['roc_auc']:.2f})")

axes[1].plot([0, 1], [0, 1], color="gray", linestyle="--")
axes[1].set_title("ROC Curve Comparison - Deep Learning Models")
axes[1].set_xlabel("False Positive Rate")
axes[1].set_ylabel("True Positive Rate")
axes[1].legend(loc="lower right")

plt.tight_layout()
plt.show()
```
